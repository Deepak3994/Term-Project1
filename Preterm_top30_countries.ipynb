```python
import pandas as pd
import numpy as np
import plotly.plotly as py
```

```python
pretop30 = pd.read_csv('Preterm_top30.csv')
pretermtop30 = pretop30[['CountryName','Preterm ']]
preterm = pretermtop30.rename(columns={'CountryName':'Country'})
preterm = preterm.sort_values(by='Country')
preterm = preterm.reset_index(drop=True)
preterm
```

```python
latlon = pd.read_csv('latitude_longitude.csv')
latlonsorted = latlon.sort_values(by='Country      ')
latlonsorted = latlonsorted[['Country      ','lat','lon']]
latitude_longitude = latlonsorted.rename(columns={'Country      ':'Country'})

latitude_longitude = latitude_longitude.reset_index(drop=True)
latitude_longitude
```

```python
remove_spaces = [x.strip().replace(' ', ' ') for x in latitude_longitude.Country]
remove_spaces = pd.DataFrame(remove_spaces)
formated = remove_spaces.rename(columns={0:"Country1"})
#formated
```

```python
new_frame = pd.concat([latitude_longitude,formated],axis=1)
new_frame = new_frame[['Country1','lat','lon']]
#new_frame
```

```python
country_list =list(preterm.Country)
country_latlon = new_frame.loc[new_frame['Country1'].isin(country_list)]
country_latlon = country_latlon.reset_index(drop=True)
country_latlon
```

```python
preterm_final = pd.concat([preterm,country_latlon],axis=1)
final = preterm_final[['Country1','Preterm ','lat','lon']]
final.to_csv('Preterm_with_latlon.csv')
```

```python
df = pd.read_csv('Preterm_with_latlon.csv')
df.head()

df['text'] = df['Country1'] + '<br>Percentage_value ' + (df['Preterm ']).astype(str)

scl = [ [0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
    [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"] ]

data = [ dict(
        type = 'scattergeo',
        locationmode = 'IS0-3',
        lon = df['lon'],
        lat = df['lat'],
        text = df['text'],
        mode = 'markers',
        marker = dict( 
            size = 8, 
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'square',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
            colorscale = scl,
            cmin = 0,
            color = df['Preterm '],
            cmax = df['Preterm '].max(),
            colorbar=dict(
                title="Percentage Value for 2015 "
            )
        ))]

layout = dict(
        title = 'Deaths of children under-five years of age due to Preterm<br>(Top 30 countries)',
        colorbar = True,   
        geo = dict(
            scope='World',
            projection=dict( type='Distance Preserving' ),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5        
        ),
    )

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='Cause of death due to Preterm' )
```
